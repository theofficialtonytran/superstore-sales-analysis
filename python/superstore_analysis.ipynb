```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/superstore_cleaned.csv', encoding='latin-1')

# Clean column names to lowercase with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')

# Discount vs Profit scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='discount', y='profit', alpha=0.4, hue='category')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1)
plt.title('Discount Rate vs Profit by Category')
plt.xlabel('Discount Rate')
plt.ylabel('Profit')
plt.tight_layout()
plt.savefig('../images/discount_vs_profit.png', dpi=150)
plt.show()
```
